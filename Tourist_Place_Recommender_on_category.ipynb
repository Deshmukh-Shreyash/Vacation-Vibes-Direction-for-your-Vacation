```python
tourist1['Place_Description'] = tourist1['Place_Description'].apply(lambda x:x.split())
tourist1
```

```python
tourist1['Place_City'] = tourist1.Place_City.apply(lambda x: x[0:].split(','))
tourist1.head()
```

```python
tourist1['Place_Category'] = tourist1.Place_Category.apply(lambda x: x[0:].split(','))
tourist1.head()
```

```python
tourist1['tags'] = tourist1['Place_Description'] + tourist1['Place_City'] + tourist1['Place_Category']
```

```python
tourist1
```

```python
new = tourist1.drop(columns=['Place_City','Place_Description'])
new
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
new['Place_Category'] = new['Place_Category'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=200,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
tourist1[tourist1['Place_Name'] == 'Prachin Konkan'].index[0]
```

```python
def recommend(place):
    city = []
    index = new[new['Place_Category'] == place].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(tourist1.iloc[i[0]].Place_Name)
        city.append(tourist1.iloc[i[0]].Place_City)
    
    return city
```

```python
recommend('Beach')
```

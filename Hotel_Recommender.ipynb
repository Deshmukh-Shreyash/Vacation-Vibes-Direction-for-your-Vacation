```python
def recommend(city,stars,price):
    new = hotels1[(hotels1['Hotel_City'] == city)]
    new1 = new[(new['Hotel_Stars'] >= stars)]
    index = new1[new1['Hotel_Price'] <= price].sort_values('Hotel_Rating',ascending=False)
    names = index['Hotel_Name'].tolist()
    
    for i in names[0:5]:
        print(i)
    return names[0:5]
```

```python
recommend('Pune',4,5000)
```

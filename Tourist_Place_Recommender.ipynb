```python
popular_df = num_rating_df.merge(avg_rating_df,on='Place_Category')
popular_df
```

```python
tourist1 = tourist[['Place_Name' ,'Place_City', 'Place_Category', 'Place_Description', 'Place_Rating']]
tourist1
```

```python
tourist1['Place_Name'].duplicated().sum()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
tourist1
```

```python
tourist1['Place_Category'] = tourist1['Place_Category'].apply(lambda x:x.split())
```

```python
tourist1
```

```python
tourist1['Place_Category'] = tourist1['Place_Category'].apply(collapse)
```

```python
tourist1['Place_Category'] = tourist1['Place_Category'].apply(lambda x: "".join(x))
tourist1.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=100,stop_words='english')
```

```python
vector = cv.fit_transform(tourist1['Place_Category']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
tourist1[tourist1['Place_Name'] == 'Prachin Konkan'].index[0]
```

```python
def recommend(place):
    index = tourist1[tourist1['Place_Name'] == place].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(tourist1.iloc[i[0]].Place_Name)
```

```python
recommend('Prachin Konkan')
```

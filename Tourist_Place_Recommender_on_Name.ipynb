```python
import numpy as np
import pandas as pd
```

```python
tourist = pd.read_csv("Final Tourist Places.csv")
tourist
```

```python
tourist.isnull().sum()
```

```python
tourist = tourist.dropna(subset=['Place_Rating','Place_Description'])
tourist
```

```python
tourist2 = tourist[['Place_Name' ,'Place_City', 'Place_Category', 'Place_Description', 'Place_Rating']]
tourist2
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
tourist2['Place_Category'] = tourist2['Place_Category'].apply(lambda x:x.split())
```

```python
tourist2
```

```python
tourist2['Place_Category'] = tourist2['Place_Category'].apply(collapse)
tourist2.head()
```

```python
tourist2['Place_Category'] = tourist2['Place_Category'].apply(lambda x: "".join(x))
tourist2.head()
```

```python
# tourist2['Place_City'] = tourist2['Place_City'].apply(collapse)
```

```python
# tourist2['Place_City'] = tourist2['Place_City'].apply(lambda x: "".join(x))
# tourist2.head()
```

```python
tourist2['Place_Description'] = tourist2['Place_Description'].apply(lambda x:x.split())
```

```python
tourist2
```

```python
# tourist2['Place_Category'] = tourist2['Place_Category'].apply(lambda x:x.split())
```

```python
tourist2['Place_Category'] = tourist2.Place_Category.apply(lambda x: x[0:].split(','))
tourist2.head()
```

```python
tourist2['Place_City'] = tourist2.Place_City.apply(lambda x: x[0:].split(','))
tourist2.head()
```

```python
tourist2['tags'] = tourist2['Place_Category'] + tourist2['Place_Description'] + tourist2['Place_City']
```

```python
tourist2
```

```python
new = tourist2.drop(columns=['Place_City','Place_Category','Place_Description'])
new
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new[new['Place_Name'] == 'Prachin Konkan'].index[0]
```

```python
def recommend(place):
    index = new[new['Place_Name'] == place].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].Place_Name)
```

```python
recommend('Shree Ganpatipule Mandir')
```
